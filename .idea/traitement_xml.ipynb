```python
import xml.etree.ElementTree as ET
import json

# Charger le fichier XML
tree = ET.parse('../corpus/IndexPersonnes.xml')
root = tree.getroot()

# Extraire tous les xml:id (parcours de tous les person)
xml_ids = []
for person in root.iter('{http://www.tei-c.org/ns/1.0}person'):
    xml_id = person.get('{http://www.w3.org/XML/1998/namespace}id')
    if xml_id:
        xml_ids.append(xml_id)

# Convertir en JSON
print(json.dumps(xml_ids, indent=2, ensure_ascii=False))

with open('output.json', 'w', encoding='utf-8') as f:
    json.dump(xml_ids, f, indent=2, ensure_ascii=False)
```
