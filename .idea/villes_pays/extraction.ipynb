```python
import json
import glob
import os

# Chemins des dossiers et fichiers
input_dir = 'data/entry_architecture'
output_cities = 'villes.json'
output_countries = 'pays.json'

def process_architecture_data():
    villes = set()
    pays = set()

    # Recherche de tous les fichiers .json dans le dossier spécifié
    search_path = os.path.join(input_dir, '*.json')
    files = glob.glob(search_path)

    if not files:
        print(f"Aucun fichier JSON trouvé dans {input_dir}")
        return

    for file_path in files:
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                
                # Extraction sécurisée des données
                place = data.get('place', {})
                city = place.get('city')
                country = place.get('country')

                if city:
                    villes.add(city)
                if country:
                    pays.add(country)
        except Exception as e:
            print(f"Erreur lors de la lecture de {file_path}: {e}")

    # Tri alphabétique et conversion en listes
    sorted_villes = sorted(list(villes))
    sorted_pays = sorted(list(pays))

    # Sauvegarde dans les fichiers JSON respectifs
    with open(output_cities, 'w', encoding='utf-8') as f:
        json.dump(sorted_villes, f, ensure_ascii=False, indent=4)

    with open(output_countries, 'w', encoding='utf-8') as f:
        json.dump(sorted_pays, f, ensure_ascii=False, indent=4)

    print(f"Traitement terminé : {len(sorted_villes)} villes et {len(sorted_pays)} pays extraits.")

if __name__ == "__main__":
    process_architecture_data()
```
